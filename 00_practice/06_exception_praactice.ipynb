### 문제 1. 안전한 정수 입력 함수

사용자에게 숫자를 입력받는 함수 `input_int(prompt)`를 작성하세요.

1. 정수가 아니면 `"정수를 입력하세요."`를 출력하세요.
2. 올바른 정수를 입력할 때까지 반복하세요.
3. 정상 입력 시 `int` 값을 반환하세요.

```python
def input_int(prompt):
    while True:
        user_input = input(prompt)
        
        try:
            result = int(user_input)
            return result
            
        except ValueError:
            print("정수를 입력하세요.")

input_int(12)
```

### 문제 2. 나눗셈 계산기

두 정수를 입력받아 `a / b` 결과를 출력하세요.

1. 0으로 나누면 `"0으로 나눌 수 없습니다."`를 출력하세요.
2. 정수가 아닌 입력이면 `"정수만 입력하세요."`를 출력하세요.
3. 어떤 경우든 프로그램이 종료되지 않고 다시 입력받도록 작성하세요.

```python
def divide():
    while True:
        try:
            input_a = input("a를 입력하세요")
            input_b = input("b를 입력하세요")

            a, b = int(input_a), int(input_b)

            result = a / b
            print(f"결과: {result}")
            break
        except ZeroDivisionError:
            print("0으로 나눌 수 없습니다.")
            
        except ValueError:
            print("정수만 입력하세요.")

    
divide()
```

### 문제 3. 파일 읽기 예외 처리

`students.txt` 파일을 읽어 학생 점수의 평균을 출력하세요.

조건

1. 파일이 없으면 `"학생 데이터 파일이 없습니다. 먼저 생성하세요."`를 출력하세요.
2. 파일이 있으면 평균을 출력하세요.
3. 예외 처리는 `try / except / else` 구조로 작성하세요.

```python
def student_score_calc():
    try:
        with open("students.txt", "r", encoding="utf-8") as f:
            scores = [int(line.strip()) for line in f if line.strip()]

    except FileNotFoundError:
        print("학생 데이터 파일이 없습니다. 먼저 생성하세요.")

    else:
        average = sum(scores) / len(scores)
        print(f"학생 점수 평균: {average}")

student_score_calc()   
```

### 문제 4. 사용자 정의 예외로 규칙 강제

점수 입력 규칙: **0~100 사이 정수만 허용**

1. 규칙을 위반하면 `InvalidScoreError` 예외를 발생시키세요.
2. main 로직에서 해당 예외를 잡아 `"점수는 0~100 사이여야 합니다."`를 출력하세요.
3. 올바른 점수를 입력할 때까지 반복하세요.

```python
class InvalidScoreError(Exception):
     """점수 범위(0~100) 위반을 나타내는 사용자 정의 예외"""

def validate_score(score):
    if not isinstance(score, (int,float)):
        raise TypeError
    if not (0 <= score <= 100):
        raise InvalidScoreError

def score_input_system():
    while True:
        try:
            score = input("0~100 사이의 정수를 입력하세요.")
            print(f"점수:{score}")
            break
        except (InvalidScoreError, TypeError):
            print("점수는 0~100 사이여야 합니다.")

    
score_input_system()
```
