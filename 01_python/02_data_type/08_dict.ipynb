# 딕셔너리 (Dictionary)
- **키(key)**와 **값(value)**으로 이루어진 자료구조
- 순서보다는 이름(key)을 통해 데이터를 빠르게 찾는 것이 목적

## 딕셔너리의 특징
- 키는 중복 될 수 없으며 변하지 않는 타입으로 사용

````python
#생성
teacher = {'name':'tiger','team':'ohgiraffers','age':40}

#키의 제약: 리스트는 키가 될 수 없지만 튜플은 가능
# invalid_dict = {['a','b']:10}
invalid_dict = {('a','b'):10}

#중복 키: 나중에 선언 된 값이 이전 값을 덮어씀
dup_dict = {'A' : 1,'A' : 2}
print(dup_dict)
````

## 데이터 조회 및 변경

````python
ogiraffers = {'Tiger':'DB','Bear':'Python','Gorilla':'Spring'}

#방법 1. 대괄호 사용 (직관적이지만 위험할 수 있음)
print(ogiraffers['Tiger'])
# print(ogiraffers['Fox']) # KeyError 발생

# 방법 2. get() 메소드 사용 (안전한 프로그래밍)
print(ogiraffers.get('Tiger'))
print(ogiraffers.get('Fox')) # None 반환 (에러 없음)
print(ogiraffers.get('Fox','미배정'))

# 데이터 추가 및 수정
ogiraffers['Fox'] = 'React' # 새로운 키-값 추가
ogiraffers['Tiger'] = 'Java' #기존 키의 값을 수정

ogiraffers
````

## 주요 메소드

````python
# 키 값 쌍 추출
print(ogiraffers.keys())    #키 목록
print(ogiraffers.values())  #값 목록
print(ogiraffers.items())   #(키, 값) 튜플 쌍 목록
````

## 딕셔너리 목사 (주의!)
파이썬의 .copy()는 **얕은 복사(shallow copy)**이다

### 불변 vs 가변
- 불변 객체(숫자, 문자열, 튜플): 값을 수정하면 아예 새로운 값을 만들어 갈아 끼우기 때문에 복사본에 영향을 주지 않음
- 가변 객체(리스트, 딕셔너리): 내부의 값을 수정해도 '주소'는 그대로이기 때문에 복사본도 그 변화를 그대로 보게됨

````python
import copy
#얕은 복사
a = {'user': {'name':'우별림'},'score':100}
b = a.copy()

a['user']['name']='김별림'  # 내부 딕셔너리 수정(주소 공유 중)
a['score'] = 90            # 바깥쪽 숫자 수정 (새로운 숫자로 교체됨)

print(f'얕은 복사 b: {b}')

#깊은 복사 -> 내부의 모든 객체까지 완전히 새로 복제
c = {'user': {'name':'우별림'},'score':100}
d = copy.deepcopy(c)

c['user']['name'] = '김별림'
c['score'] = 90

print(f'깊은 복사 d: {d}')  # 원본을 수정해도 복사본은 독립적임
````
