# 조건문 if
- 특정 조건의 참/거짓에 따라 실행 경로를 결정함
- 들여쓰기가 코드 범위를 결정하므로 주의 필요함

## 단일 if문
- 특정 조건이 충족 될 때만 실행해야 하는 경우 사용

```python
#90점 이상인 경우에만 커피 쿠폰 지급
score = int(input('시험 점수 입력: '))

if score >= 90:
    print('커피 쿠폰을 드립니다.')

print('수고하셨습니다.')
```

## if-else문
- 조건이 참일 때와 거짓일 때 각각 실행할 명령이 있는 경우 사용

```python
# 합격/불합격 판독
score = int(input('시험 점수 입력: '))
if score >= 60:
    print('합격입니다.')
else:
    print('불합격입니다.')

print('수고하셨습니다.')
```

## if-elif-else 문
- 여러 조건을 순차적으로 확인해야하는 경우 사용
- 상단 조건이 만족되면 하단 조건은 검사하지 않음

```python
# 점수별 학점 등급 부여
score = int(input('시험 점수 입력: '))
grade = None

if score >= 90:
    grade = 'A'
elif score >= 80:
    grade = 'B'
elif score >= 70:
    grade = 'C'
elif score >= 60:
    grade = 'D'
else:
    grade = 'F'

print(f'수고하셨습니다. 당신의 점수는 {score}점이고, 등급은 {grade}입니다.')
```
