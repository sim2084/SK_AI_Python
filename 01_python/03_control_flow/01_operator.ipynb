# Operator (연산자)
- 데이터를 계산하고 비교하는 기초 기호
- 우선 순위가 복잡 할 때는 괄호 () 사용이 권장된다.

## 산술 및 논리 연산자 우선 순위
1. () : 괄호
2. ** : 제곱거듭
3. *,/,//,% : 곱셈, 나눗셈, 몫, 나머지
4. +,- : 덧셈, 뺄셈
5. ==,!=,>,<,>=,<= : 비교 연산자
6. not > and > or : 논리 연산자

````python
#산술 연산자 활용
num = 7
print(f'num은 짝수인가? {num % 2 == 0}')
print(f'num은 3의 배수인가? {num % 3 == 0}')
````

````python
# 논리 연산자 활용
# and : 모두 참이어야 함
# or : 하나라도 참이면 참
# not : 참이면 거짓이고, 거짓이면 참으로

# 놀이기구 탑승 조건 : 나이가 10세 이상이며 키가 130cm 이상이어야 함
age = 12
height = 135
can_ride = age >= 10 and height >= 130
print(f'놀이기구 탑승 가능 여부: {can_ride}')

#무료 입장 대상자 확인 : 나이가 65세 이상이거나, 7세 미만인 경우
is_free = age < 7 or age >= 65
print(f'무료 입장 대상 여부: {is_free}')

#서비스 점검 상태 확인
is_under_maintenance = False
can_use_service = not is_under_maintenance # 점검 중이 아니라면 사용 가능
print(f'서비스 이용 가능 여부:{can_use_service}')
````

````python
# 조건부 표현식 (참일때 값 if 조건 else 거짓일때 값)
n = int(input('정수를 입력하세요: '))
result = '양수' if n > 0 else '음수'
result
````

````python
name = input('이름 압력: ')
response = f'{name}님' if name else '이름없음'
response
````
