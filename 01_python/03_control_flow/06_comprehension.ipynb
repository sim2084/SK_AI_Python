# 컴프리헨션 (comprehension,내포)
- 반복문과 조건문을 결합하여 리스트나 딕셔너리를 한 줄로 간결하게 생성하는 방법
- 코드가 짧아지고 가독성이 좋아짐

## 리스트 내포
- 기본 구조: [표현식 for 변수 in 반복가능객체 if 조건]

```python
# 1부터 10까지의 제곱값 리스트 생성
lst_old = []
for n in range(1,11):
    lst_old.append(n ** 2)

print(lst_old)

lst_new = [n ** 2 for n in range(1,11)]

print(f'제곱 리스트: {lst_new}')

# 1~100 사이의 짝수만 추출
even_lst = [n for n in range(1,101) if n % 2 == 0]

print(f'짝수 리스트 {even_lst}')
```

## 데이터 필터링 연습(isinstance)
- 리스트 내에서 특정 자료형(예: 정수)만 골라내는 유용한 기법

```python
#다양한 데이터 타입이 섞인 리스트
mixed_data = [1,'ㅋㅋㅋ',123.456,33,'ㅎㅎ',0.123]
#isinstance(객체,int): 객체가 정수이면 True를 반환
int_only = [n for n in mixed_data if isinstance(n,int)]
print(f'필터링 결과: {int_only}')
```

## 중첩 반목문 내포
- 이중 for문도 한 줄로 표현 가능

```python
coords = [(row,col) for row in range(0,3) for col in range(0,5)]
print(f'생성한 좌표: {coords}')
```

## 딕셔너리 내포
- {키: 값 for ...} 형태로 딕셔너리를 생성

```python
#숫자: 제곱값 딕셔너리
square_dict = {k: k**2 for k in range(1,6)}
print(f'제곱 사전: {square_dict}')

#문자열 리스트를 이용해 {단어:길이} 딕셔너리 만들기. 단, 3글자 이상만.
words = ['hello','hi','monday']
word_len_dict = {word:len(word) for word in words if len(word) >= 3}
print(f'단어 길이 사전: {word_len_dict}')
```

## 고급 활용 : zip()과 unpacking() 연산자(*)
- 데이터를 조합하거나 형변환할 때 강력한 기능을 발휘

```python
names = ['홍길동','신사임당','이순신']
scores = [100,85,95]

#두 리스트를 묶어서 딕셔너리로 변환
user_data = dict(zip(names,scores)) # 두 개의 순화 가능한 데이터를 인덱스 순서에 맞게 짝지어줌
print(f'묶기 결과: {user_data}')

# 리스트의 포장지 뜯기
numbers = [1,2,3]
print(f'포장 듣기 전: {numbers}')
print('포장 듣은 후: ', *numbers)

pairs = [('Apple',5000),('Banana',3000),('Kiwi',8000)]
# 1. *pairs -> 리스트 대괄호 제거
# 2. zip이 개별 튜플을 다시 세로로 묶음
items, prices = zip(*pairs)

print(f'상품명만 분리: {items}')
print(f'가격만 분리: {prices}')
```
