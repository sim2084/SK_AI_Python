# 다형성
- 같은 이름의 메소드를 호출하지만 객체의 타입에 따라 서로 다른 결과가 나오는 능력

## 메소드 오버라이딩 (Method Overiding)
- 부모 메소드를 자식이 자신 입맛에 맞게 다시 정의하는 다형성의 핵심 기술

```python
class Animal:
    def speak(self):
        return '...'

class Dog(Animal):
    def speak(self):
        return '멍멍'

class Cat(Animal):
    def speak(self):
        return '야옹'

animals = [Dog(),Cat(),Dog(),Cat(),Animal()]
for animal in animals:
    print(f'{type(animal).__name__}:{animal.speak()}')
```

## 덕 타이핑(Duck Typing)
"오리처럼 걷고 꽥꽥거리면 오리다." => 상속 관계가 아니더라도 메소드 이름만 같으면 동일하게 취급하는 파이선의 유연한 특징

```python
class Tiger:
    def speak(self):
        return '어흥'

class Car:
    def speak(self):
        return '빵빵'

def make_it_speak(obj):
    # obj가 무엇이든 상관 없이 speak() 메소드만 있으면 실행 가능
    print(obj.speak())


make_it_speak(Tiger())
make_it_speak(Car())
```

## 추상 클래스(Abstract Class)
- 자식 클래스에서 특정 메소드를 **반드시 구현하도록 강제** 할 때 사용
- 협업 시 실수를 줄여주는 안전 장치의 역할

```python
# Abstract Base Class (ABC) 모듈 사용하기
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):     #자식은 반드시 이 메소드를 구현해야 하는 강제
        pass

class Rectangle(Shape):
    def __init__(self, w, h):
        self.w, self.h = w,h        # 가로, 세로 인스턴스 생성
    def area(self):
        return self.w*self.h

class Circle(Shape):
    def __init__(self, r):
        self.r = r                  # 반지름 인스턴스 생성
    def area(self):
        return 3.14*(self.r**2)

shapes = [Rectangle(3,5), Circle(4)]
for shape in shapes:
    print(f'넓이 : {shape.area()}')
```

## 다형성의 실무적 장점
1. 유지보수성 (OCP: Open-Closed Principle)
- 확장에는 열려 있고 수정에는 닫혀있어야 한다는 원칙
- 예를 들어 ApplePay와 같은 결제 수단이 추가된다면? ApplePay 클래스 추가, 실제 결제 처리 함수는 수정 불필요
2. 코드의 단순화
- 만약 다형성이 적용되지 않을 경우 함수 내부에서 if instance(method,KakaoPay)처럼 객체를 검색해야 하지만 다형성이 적용 되어 어떤 객체이든 상관 없고 pay 하라는 명령만 내리면 된다.
3. 결합도 낮추기
- 결합도 : 특정 클래스의 변화가 다른 곳에 영향을 주는 정도
- process_payment 함수는 구체적 카드, 카카오 페이 등에 의존하지 않고 추상적인 payment의 약속에만 의존한다. 따라서 결제 수단이 바뀌거나 늘어나도 핵심 로직은 영향 받지 않는다.

```python
class Payment(ABC):
    @abstractmethod
    def pay(self, amount):
        pass

class Creditcard(Payment):
    def pay(self, amount):
        print(f'신용카드로 {amount} 결제하는 로직')

class kakaoPay(Payment):
    def pay(self, amount):
        print(f'카카오 페이로 {amount} 결제하는 로직')

def process_payment(method, amount):
    method.pay(amount)

# 결제 수단이 무엇이든 동일한 방식으로 처리 가능
process_payment(Creditcard(), 15_000)
process_payment(kakaoPay(), 20_000)
```
